```python
#using the second time data from experiment at 04/11/2021
f = open("secondExp04112021.txt", "r")

Lines = f.readlines()
 
count = 0
xIndex=[]
yIndex=[]
# Strips the newline character
for line in Lines:
    count += 1
    #print(line.strip())
    xIndex.append(float(line.strip()[0:4]))
    #print(line.strip())
    yIndex.append(int(line.strip()[-5:]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.title("pwm for valve")
plt.plot(xIndex, yIndex)
plt.show()
```
